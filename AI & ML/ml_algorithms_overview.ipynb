## 1. Introduction to Machine Learning
Machine Learning (ML) is a field of artificial intelligence that enables computers to learn from data and make predictions or decisions. ML algorithms are generally categorized as supervised or unsupervised.

## 2. Supervised Learning Algorithms
Supervised learning uses labeled data to train models. Common tasks: regression and classification.

### 2.1 Linear Regression
Linear Regression is used for predicting a continuous target variable based on one or more input features.

```python
# Linear Regression Example
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate synthetic data
X = np.random.rand(100, 1) * 10
y = 2.5 * X.flatten() + np.random.randn(100) * 2

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

plt.scatter(X, y, label='Data')
plt.plot(X, y_pred, color='red', label='Linear Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression Example')
plt.show()
```
